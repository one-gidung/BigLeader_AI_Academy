```python
import tensorflow
```

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = Sequential()
model.add(Dense(1, input_dim=13, activation='linear'))
```

```python
# data: 10000, batch_size: 100, epochs: 100
```

```python
# otms = optimizers.SGD(learning_rate=0.01) # no training
# otms = optimizers.Adagred(learning_rate=0.01) # okay
# otms = optimizers.Adamax(learning_rate=0.01) # okay
# otms = optimizers.Ftrl(learning_rate=0.01) # okay
otms = optimizers.Nadam(learning_rate=0.01) # okay
# otms = optimizers.RMSpropp(learning_rate=0.01) # problem

model.compile(optimizer=otms, loss='mse', metrics=['accuracy'])
```

```python
model.fit(train_data, train_targets, epochs=100, batch_size=100, shuffle=False)
```

```python
def build_model():
    model = Sequential()
    model.add(Dense(1, input_dim=train_data.shape[1], activation='linear'))
    model.compile(optimizer='nadam', loss='mse', metrics=['accuracy'])
    return model
```

```python
# 교차 검증

k = 4
num_val_samples = len(train_data) // k
num_epochs = 100
all_scores = []
for i in range(k):
    print("처리중인 폴드: #", i)
    val_data = train_data[i * num_val_samples: (i+1)*num_val_samples]
    val_targets = train_targets[i * num_val_samples: (i+1)*num_val_samples]
    
    partial_train_data = np.concatenate(
        [train_data[:i*num_val_samples],
        train_data[(i+1) * num_val_samples:]], axis=0
    )
    
    partial_train_targets = np.concatenate(
        [train_targets[:i*num_val_samples],
        train_targets[(i+1) * num_val_samples:]], axis=0
    )
    model = build_model()
    model.fit(partial_train_data, partial_train_targets, epochs=num_epochs, batch_size=10)
    
    val_mse, accuracy = model.evaluate(val_data, val_targets)
    all_scores.append(val_mse)
    print('all_scores: ', all_scores)
```

```python
np.mean(all_scores)
```

```python
model.evaluate(test_data, test_targets, batch_size=10)
```

```python
# 1차 epoch 100: validation 평균: 424.992, test score: 436.596
# 2차 epoch 500: validation 평균: 94.775, test score: 91.850
```
