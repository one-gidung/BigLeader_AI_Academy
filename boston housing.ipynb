```python
from tensorflow.keras.datasets import boston_housing
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras import regularizers
```

```python
data = boston_housing.load_data()
data
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
# 정규화
mean = x_train.mean()
std = x_train.std()
x_train -= mean
x_train /= std

x_test -= mean
x_test /= std
```

```python
x_train.shape
```

```python
# 모델 설계
model = Sequential()
model.add(Dense(1, input_dim=13, activation='linear'))
```

```python
# 옵티마이저를 나중에 다른 걸 사용할 수도 있기 때문에 변수로 하나 만들어 준다.
opt = optimizers.RMSprop(learning_rate=0.01)
```

```python
model.compile(optimizer=opt, loss='mse', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=1, shuffle=False)
```

```python
# 모델 평가
# loss 값이 작을 수록 좋다. mse가 작다는 뜻!
model.evaluate(x_test, y_test)
```
