```python
import tensorflow
```

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
x_train[1]
```

```python
y_train[1]
```

```python
import pandas as pd
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

df = pd.DataFrame(x_train, columns=columns)
df.head()
```

```python
df_test = pd.DataFrame(x_test, columns=columns)
df_test.head()
```

```python
df.info()
```

```python
df_test.info()
```

```python
y_train
# 10의 단위 이므로 standard scaler로 값 보정
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
df = ss.fit_transform(df)
df
```

```python
df_test = ss.transform(df_test)
df_test
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_dim = 13, activation = 'linear'))
```

```python
sgd = optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=sgd, loss = 'mse', metrics=['acc'])
```

```python
model.fit(df, y_train, epochs=100, batch_size = 1, shuffle=False)
```

```python
df_test
```

```python
y_predict = model.predict([df_test])
y_predict
```

```python
model.evaluate(df_test, y_test, verbose=1, batch_size = 100)
```
